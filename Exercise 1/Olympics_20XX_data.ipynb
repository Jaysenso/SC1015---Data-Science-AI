```python
import numpy as np
import pandas as pd
```

```python
olympics_20XX = {}
for x in range(2000,2020,4):
    current_data =  pd.read_html('https://en.wikipedia.org/wiki/'+str(x) +'_Summer_Olympics_medal_table')
    if (current_data):
        olympics_20XX[str(x)] = (current_data[2])
```

```python
olympics_20XX.items()
```

```python
olympic_top_20_data = {}
for year, data in olympics_20XX.items():
    olympic_top_20_data[year] = data.head(n = 20)
    
for x,y in olympic_top_20_data.items():
    print('Year: ',x)
    print(y)
    print()
```
